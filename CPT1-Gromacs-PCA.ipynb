```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
df = pd.read_excel('PCA.xlsx')
```

```python
#Apo CPT1

apo = df.loc[0:505,['PC1','PC2']]


apo_plot = sns.jointplot(x='PC1',y='PC2',data=apo, color = 'silver', kind='kde')
apo_plot.ax_marg_x.set_xlim(-15, 12)
apo_plot.ax_marg_y.set_ylim(-12, 10)
   
   

fig1 = plt.gcf()
plt.draw()
fig1.savefig('Apo_CPT1-PCA.pdf', dpi=600)
```

```python
#Alpha

Alpha = df.loc[505:1010,['PC1','PC2']]


Alpha_plot = sns.jointplot(x='PC1',y='PC2',data=Alpha, color = '#f2b77c', kind='kde')


Alpha_plot.ax_marg_x.set_xlim(-15, 12)
Alpha_plot.ax_marg_y.set_ylim(-12, 10)


fig2 = plt.gcf()
plt.draw()
fig2.savefig('Alpha_CPT1-PCA.pdf', dpi=600)
```
